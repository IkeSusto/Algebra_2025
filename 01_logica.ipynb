### Negação:$\quad \sim p$



> O ideia da proposição "não p" é negar a proposição p, ou seja, ser contrária/oposta a proposição original. 

- *lê-se: "não p"*
- $\pink\sim \blue{p}:\; \pink{não} \; \blue{p}$ 

#### Exemplos:
- $\quad p$ : Gosto de gatos
- $\sim p$ : ***Não*** gosto de gatos
- $\quad q$ : 3 e 5 dividem 10
- $\sim q$ : 3 e 5 ***não*** dividem 10
- $\quad r$ : O professor de álgebra é legal
- $\sim r$ : O professor de álgebra é ***chato***

Obs.: Em $r$ consideramos que o oposto de legal é chato e vice versa.

Note a proposição $\;p$ só será <span style="color:#77dd77">Verdadeira</span> se sua respectiva negação $\;\sim p$ for <span style="color:#ff6961">Falsa</span>. Ainda, $\;p$ só será <span style="color:#ff6961">Falsa</span> se sua respectiva negação $\;\sim p$ for <span style="color:#77dd77">Verdadeira</span>. O mesmo raciocínio vale para $q$ e $r$.

Os cenários em que $p$ e $\;\sim p$ são iguais — ambas <span style="color:#77dd77">Verdadeiro</span> ou ambas <span style="color:#ff6961">Falsas</span> — são logicamente impossíveis, independentemente de quem seja p.
- $p$ <span style="color:#77dd77">Verdadeira</span> e $\;\sim p$ <span style="color:#77dd77">Verdadeira</span>: Gosto e não gosto de gatos
- $r$ <span style="color:#ff6961">Falsa</span> e $\;\sim r$ <span style="color:#ff6961">Falsa</span>: O professor de álgebra não é legal e não é chato

Essas frases não fazem o menor sentido logicamente. 

Note que $s$: "Gosto um pouco de gatos" não é mesma coisa que "Gosto e não gosto de gatos". Repetindo a lógica acima para $s$ teríamos:
- $s$ <span style="color:#77dd77">Verdadeira</span> e $\;\sim s$ <span style="color:#77dd77">Verdadeira</span>: Gosto um pouco e não gosto um pouco de gatos. Continua sem sentido lógico.
 

#### Tabela verdade
| $p$ | $\sim p$ |
|:---:|:--------:|
| <span style="color:#77dd77">V</span> | <span style="color:#ff6961">F</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#77dd77">V</span> |

*Regra para preencher a tabela:*
1. Escrevo todas as possibilidades de $p$
1. $\;\sim p$ só é <span style="color:#77dd77">Verdadeiro</span> quando $p$ for <span style="color:#ff6961">Falso</span>.
1. Todas as outras linhas são <span style="color:#ff6961">Falsas</span>

### Conjunção:$\quad p \land q$



> O nome conjunção vem do fato de que para $p \land q$ ser <span style="color:#77dd77">verdadeira</span> preciso que $p$ e $q$ sejam, em conjunto, ao mesmo tempo, <span style="color:#77dd77">verdadeiras</span>. Ideia de união, preciso das duas.

Elas juntas definem a conjunção. Garantir que só uma das duas proposições é verdadeira, não é suficiente para concluir que $p \land q$ também será. Mas se ambas forem verdadeiras, em conjunto, é possível garantir que a conjunção $p \land q$ também será verdadeira.

- *lê-se: "p e q"*
- $p \pink\land q\; :\; p\; \pink{e}\; q$

#### Exemplos:
- $\quad\;\; p$ : Meu é lanche gostoso.
- $\quad\;\; q$ : Meu é lanche bonito.
- $p\land q$ : Meu lanche é bonito **e** meu lanche é gostoso.

Aqui poderíamos reduzir a expressão ainda mais e dizer apenas "Meu lanche é bonito **e** gostoso". Nem sempre essa redução é simples

O lanche que comprei só pode ser bonito e gostoso ($p\land q$ <span style="color:#77dd77">Verdadeiro</span>) se ao mesmo tempo $-$ conjuntamente $-$ ele for gostoso ($p$ <span style="color:#77dd77">Verdadeiro</span>) e bonito ($q$ <span style="color:#77dd77">Verdadeiro</span>).

- $\quad\;\; r$ : Quando está frio eu tomo banho quente.
- $\quad\;\; s$ : Quando tomo banho quente eu canto no chuveiro.
- $r\land s$ : Quando está frio eu tomo banho quente **e** quando tomo banho quente canto no chuveiro.

Aqui poderíamos reduzir para "Quando está frio eu tomo banho quente **e** canto no chuveiro"

Note aqui que se $r$ for <span style="color:#ff6961">Falsa</span> a proposição $r\land s$ também será <span style="color:#ff6961">Falsa</span> independente do valor de $s$. O mesmo raciocínio vale para $s$.
- Se $r$ é <span style="color:#ff6961">F</span>, significa que em algum dia frio eu tomei um banho não quente. Note que nem preciso olhar para o restante da frase para conluir que a frase "Quando está frio eu tomo banho bem quente **e** canto no chuveiro", $r \land s$, é <span style="color:#ff6961">Falsa</span>.

#### Tabela verdade
| $p$ | $q$ | $p \land q$ |
|:---:|:---:|:-----------:|
| <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> |
| <span style="color:#77dd77">V</span> | <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#77dd77">V</span> | <span style="color:#ff6961">F</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> |

*Como preencher a tabela:*
1. Escrevo todas as possíveis combinações de $p$ e $q$ : <span style="color:#77dd77">V</span><span style="color:#77dd77">V</span>, <span style="color:#77dd77">V</span><span style="color:#ff6961">F</span>, <span style="color:#ff6961">F</span><span style="color:#77dd77">V</span> e <span style="color:#ff6961">F</span><span style="color:#ff6961">F</span>.
1. $p \pink\land q$ só é <span style="color:#77dd77">Verdadeiro</span> quando ambos $p$ $\pink e$ $q$ forem <span style="color:#77dd77">Verdadeiros</span>. Ou seja, só preencho <span style="color:#77dd77">V</span> para a coluna $p \pink\land q$  na linha <span style="color:#77dd77">V V</span>.
1. Todas as outras linhas são <span style="color:#ff6961">Falsas</span>

---


### Disjunção:   $p \lor q$



> O nome disjunção vem do fato de que para $p \lor q$ ser <span style="color:#77dd77">verdadeira</span> basta que $p$ e $q$ sejam, separadamente, <span style="color:#77dd77">verdadeiras</span>. Ideia de desunião, basta uma.

Enquanto para $``p\; e\; q"$ ser verdadeira, $p$ e $q$ precisavam ser, *juntamente*, verdadeiras; para $``p\; ou\; q"$ é ser verdadeira, basta uma das duas, "separadamente"  ser <span style="color:#77dd77">verdadeira</span>.

*lê-se: "p ou q"*

$p \pink\lor q:\; p\; \pink{ou}\; q$

#### Exemplos
- $\quad\;\; p$ : Vou ao cinema de metrô.
- $\quad\;\; q$ : Vou ao cinema de ônibus.
- $p\land q$ : Vou ao cinema de metrô ou vou ao cinema de ônibus.

Aqui poderíamos reduzir a disjunção para "Vou ao ônibus de metrô ou ônibus". Note que essa frase continua Verdadeira mesmo quando eu vou só de metrô ($p$ V e $q$ F) ou só de ônibus ($p$ F e $q$ V). Basta que uma delas seja Verdadeira para a disjunção $p\lor q$ ser Verdadeira.

#### Tabela verdade
| $p$ | $q$ | $p \lor q$ |
|:---:|:---:|:----------:|
| <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> |
| <span style="color:#77dd77">V</span> | <span style="color:#ff6961">F</span> | <span style="color:#77dd77">V</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> |

*Como preencher a tabela:*
- $p \lor q$ só é <span style="color:#ff6961">Falso</span> quando ambos $p$ $\lor$ $q$ forem <span style="color:#ff6961">Falsos</span>. Ou seja, basta procurar todas as linhas <span style="color:#ff6961">F F</span>. 
- Todas as outras linhas são <span style="color:#77dd77">Verdadeiras</span>

Pela tabela verdade podemos observar que $p\lor q$ só <span style="color:#ff6961">Falsa</span> quando ambas $p$ e $q$ forem simulataneamente falsas. Essa será a nossa leitura padrão.
- $p\land q$ só é $V$ na linha <span style="color:#77dd77">V V</span>
- $p\lor q$ só é <span style="color:#ff6961">F</span> na linha <span style="color:#ff6961">F F</span>




### Condicional:   $p \rightarrow q$

> *lê-se: "se p então q"*

$p \pink\rightarrow q:\; \pink{se}\; p\; \pink{então}\; q$

Tabela verdade
| $p$ | $q$ | $p \rightarrow q$ |
|:---:|:---:|:-----------------:|
| <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> |
| <span style="color:#77dd77">V</span> | <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> | <span style="color:#77dd77">V</span> |

*Como preencher a tabela:*
- $p \rightarrow q$ só é <span style="color:#ff6961">Falso</span> quando $p$ é <span style="color:#77dd77">Verdadeiro</span> e $q$ é <span style="color:#ff6961">Falso</span>. Ou seja, basta procurar todas as linhas <span style="color:#77dd77">V</span> <span style="color:#ff6961">F</span>. 
- Todas as outras linhas são <span style="color:#77dd77">Verdadeiras</span>



### Bicondicional:   $p \leftrightarrow q$

Podemos definir a bicondicional como sendo as duas condicionais ao mesmo tempo $(p \rightarrow q)$ e $(p \leftarrow q)$

A ideia dessa condicional é ser uma setinha de mão dupla, ela vai $(p \rightarrow q)$ $(p \leftarrow q)$. $se p então q" e "se q então p" ao mesmo tempo.

> *lê-se: se p então q e se q então p" ou "se p então q e vice versa" ou "p se e somente se q"*

Note que em todas as formas de leitura aparece um $\pink{e}$ em algum lugar, o que condiz muito com a existência do nosso conectivo $\pink\land$ na definição.

$p \pink\leftrightarrow q: (p \pink\rightarrow q)\pink\land(q \pink\rightarrow p)$
$\; \pink{se}\; p\; \pink{então}\; q\;\pink{\;e\; vice\; versa}$
$p\; \pink{se\;e\;somente\;se}\; q$

Tabela verdade
| $p$ | $q$ | $p \leftrightarrow q$ |
|:---:|:---:|:---------------------:|
| <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> | <span style="color:#77dd77">V</span> |
| <span style="color:#77dd77">V</span> | <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#77dd77">V</span> | <span style="color:#ff6961">F</span> |
| <span style="color:#ff6961">F</span> | <span style="color:#ff6961">F</span> | <span style="color:#77dd77">V</span> |

*Como preencher a tabela:*
- $p \leftrightarrow q$ só é <span style="color:#77dd77">Verdadeiro</span> quando ambos $p$ e $q$ são <span style="color:#77dd77">Verdadeiros</span> ou ambos são <span style="color:#ff6961">Falsos</span>. Ou seja, basta procurar todas as linhas <span style="color:#77dd77">V V</span> ou <span style="color:#ff6961">F F</span>. 
- Todas as outras linhas são <span style="color:#ff6961">Falsas</span>
