### Proposições



### <cp>Negação:$\quad \sim p$</cp>


> O ideia da proposição "não p" é negar a proposição p, ou seja, ser contrária/oposta a proposição original. 

Lê-se
- <cp>$\sim$</cp>$\;p\; :\;$<cp>não</cp>$\;p$ 

#### Exemplos:
- $\quad p\;:\;$ Gosto de gatos
- <cp>$\sim$</cp>$\;p\; :\;$<cp>não</cp> gosto de gatos
- $\quad q\;:\;$ 3 e 5 dividem 10
- <cp>$\sim$</cp>$\;q\; :\;$ 3 e 5 <cp>não</cp> dividem 10
- $\quad r\;:\;$ O professor de álgebra é legal.
- <cp>$\sim$</cp>$\;r\; :\;$ O professor de álgebra <cp>não</cp> é legal.

Note a proposição $\,p\,$ só será <cg>$\,verdadeira\,$</cg> se sua respectiva negação $\,\sim p\,$ for <cr>$\,falsa\,$</cr>.
Ainda, $\,p\,$ só será<cr>$\,falsa\,$</cr> se sua respectiva negação $\,\sim p\,$ for <cg>$\,verdadeira\,$</cg>. O mesmo raciocínio vale para $\,q\,$ e para $\,r\,$.

Os cenários em que $\,p\,$ e $\;\sim p\,$ são iguais — ambas <cg>$\,verdadeiras\,$</cg> ou ambas <cr>$\,falsas\,$</cr> — são logicamente impossíveis, independentemente de quem seja p. 

Exemplos:
- $(\,p\,$<cg>$\,verdadeira\,$</cg>$)$ e $(\,\sim p\,$<cg>$\,verdadeira\,$</cg>$)\,:\,$ Gosto e não gosto de gatos.
- $(\,r\,$<cr>$\,falsa\,$</cr>$)$ e $(\,\sim r\,$<cr>$\,falsa\,$</cr>$)\,:\,$ O professor de álgebra não é legal e não não é legal. Aqui um não cancela o outro, pois o oposto do oposto é o próprio. Ficamos então com: "O professor de ãlgebra não é legal e é legal".

Essas frases não fazem sentido lógico. 

Note que $\,s\,$: "Gosto um pouco de gatos" não é mesma coisa que "Gosto e não gosto de gatos". Repetindo a lógica acima para $\,s\,$ teríamos:
- $\,s\,$<cg>$\,verdadeira\,$</cg> e $\,\sim s\,$<cg>$\,verdadeira\,$</cg>: Gosto um pouco de gatos e gosto muito de gatos. Novamente chegamos em uma contradição.

| $p$ | $\sim p$ |
|:---:|:--------:|
|<cg>$\,V\,$</cg> |<cr>$\,F\,$</cr> |
|<cr>$\,F\,$</cr> |<cg>$\,V\,$</cg> |

*Regra para preencher a tabela:*
1. Escrevo todas as possibilidades de $\,p\,$
1. $\,\sim p\,$ só é <cg>$\,verdadeira\,$</cg> quando $p$ for <cr>$\,falsa\,$</cr>.
1. Todas as outras linhas são <cr>$\,falsas\,$</cr>.

### <co>Conjunção:$\quad p \land q$</co>


> O nome conjunção vem do fato de que para $p \land q$ ser $$verdadeira}$ preciso que $p$ e $q$ sejam em conjunto, ao mesmo tempo, <cg>$\,verdadeiras\,$</cg>. Ideia de união, preciso das duas.

Elas juntas definem a conjunção. Garantir que só uma das duas proposições é <cg>$\,verdadeira\,$</cg>, não é suficiente para concluir que $p \land q$ também será. Mas se ambas forem <cg>$\,verdadeiras\,$</cg>, em conjunto, é possível garantir que a conjunção $p \land q$ também será <cg>$\,verdadeira\,$</cg>.

- *lê-se: "p e q"*
- $p$ <co>$\land$</co> $q\; :\; p\;$ <co>$e$</co>$\; q$

#### Exemplos:
- $\quad\;\; p$ : Meu é lanche gostoso.
- $\quad\;\; q$ : Meu é lanche bonito.
- $p\;$<co>$\land$</co>$\;q$ : Meu lanche é bonito <co>e</co> meu lanche é gostoso.

Aqui poderíamos reduzir a expressão ainda mais e dizer apenas "Meu lanche é bonito <co>e</co> gostoso". Nem sempre essa redução é simples

O lanche que comprei só pode ser bonito e gostoso ($p\land q$ <cg>$\,verdadeira\,$</cg>) se ao mesmo tempo $-$ conjuntamente $-$ ele for gostoso ($p$ <cg>$\,verdadeira\,$</cg>) e bonito ($q$ <cg>$\,verdadeira\,$</cg>). Daí o nome conjunção.

- $\quad\;\; r$ : Quando está frio eu tomo banho quente.
- $\quad\;\; s$ : Quando tomo banho quente eu canto no chuveiro.
- $r\land s$ : Quando está frio eu tomo banho quente <co>e</co> quando tomo banho quente canto no chuveiro.

Aqui poder

íamos reduzir para "Quando está frio eu tomo banho quente <co>e</co>e canto no chuveiro"

Note aqui que se $r$ for <cr>$falsa$</cr> a proposição $r\land s$ também será <cr>$falsa$</cr> independente do valor de $s$. O mesmo raciocínio vale para $s$.
- Se $r$ é <cr>$falsa$</cr>, significa que em algum dia frio eu tomei um banho não quente. Note que nem preciso olhar para o restante da frase para conluir que a frase "Quando está frio eu tomo banho bem quente <co>e</co> canto no chuveiro", $r \land s$, é <cr>$falsa$</cr>.

1. Todas as outras linhas são <cr>$falsas$</cr>

### <cb>Disjunção:$\quad p \lor q$</cb>


> O nome disjunção vem do fato de que para $p \lor q$ ser <cg>$\,verdadeira\,$</cg> basta que $p$ e $q$ sejam, separadamente, <cg>$\,verdadeiras\,$</cg>. Basta uma, ideia de desunião, disjunção.

Enquanto para <co>$\;p\land q\;$</co> ser verdadeira, $p$ e $q$ precisavam ser *conjuntamente* <cg>$\,verdadeiras\,$</cg>; para <cb>$\;p\lor\; q\;$</cb> ser <cg>$\,verdadeira\,$</cg>, basta que $p$ e $q$ sejam *disjuntamente* <cg>$\,verdadeiras\,$</cg>.

*lê-se: "p ou q"*

$p$ <cb>$\lor$</cb> $q:\; p\;$ <cb>$ou$</cb>$\; q$

#### Exemplos
- $\quad\;\; p$ : Vou ao cinema de metrô.
- $\quad\;\; q$ : Vou ao cinema de ônibus.
- $p\land q$ : Vou ao cinema de metrô ou vou ao cinema de ônibus.

Aqui poderíamos reduzir a disjunção para "Vou ao cinema de metrô ou ônibus". Note que essa frase continua <cg>$\,V\,$</cg> mesmo quando eu vou só de metrô ($p$ <cg>$\,V\,$</cg>, $q$ <cr>$\,F\,$</cr>) ou só de ônibus ($p$ <cr>$\,F\,$</cr>, $q$ <cg>$\,V\,$</cg>.

#### Tabela verdade
| $p$ | $q$ | $p \lor q$ |
|:---:|:---:|:----------:|
| <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> | 
| <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> |
| <cr>$\,F\,$</cr> | <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> |
| <cr>$\,F\,$</cr> | <cr>$\,F\,$</cr> | <cr>$\,F\,$</cr> |

*Como preencher a tabela:*
- $p \lor q$ só é <cr>$\,F\,$F\,$</cr>, basta procurar to<cr>$\,F\,$</cr>
- Todas as outras linhas são <cg>$\,V\,$</cg>

Pela tabela verdade podemos observar que $p\lor q$ só é <cr>$\,F\,$</cr> quando $p$ e $q$ forem simulataneamente <cr>$\,F\,$</cr>
- $p\land q$ só é <cg>$\,V\,$</cg> na linha <cg>$\,VV\,$</cg>
- $p\lor q$ só é <cr>$\,F\,$</cr> na linha <cr>$\,FF\,$</cr>

### <cv>Condicional:$\quad p \rightarrow q$</cv>

> O nome condicional já deixa claro que estamos falando de uma relação em que uma coisa é condição para outra. Se eu garantir uma certa coisa (condição) posso garantir uma outra coisa (conclusão).

- Ter mais de 18 anos é condição suficiente para poder comprar bebidas alcóolicas.
    - Se tenho sou maior de idade então posso comprar bebidas. 
- Estar grávida é condição suficiente para poder usar o assento preferencial em transportes públicos.
    - Se estou grávida então posso usar o assento preferencial.
- Estar matriculado no CPE é condição necessária para assistir as aulas do cursinho.
    - Se assisto aulas do cursinho então estou matriculado no CPE.
- Ter asas é condição necessária para poder voar.
    - Se posso voar então tenho asas.

Avaliando as frases acima, teríamos <cg>$\,V\,V\,V\,$</cg><cr>$F\,$</cr>

*lê-se: "se p então q"*

$p$ <cv>$\rightarrow$</cv> $q:\;$ <cv>$se$</cv>$\; p\;$ <cv>$então$</cv>$\; q$

Existem também as seguintes leituras para $p \rightarrow q$ como explicitado nas frases acima.
- "p é condição <cv>suficiente</cv> para q"
- "q é condição <cv>necessária</cv> para p"

#### Exemplos

- $\quad\;\; p$ :  $3\leq5$
- $\quad\;\; q$ :  $\dfrac{3}{5}\leq1$
- $p\land q$ :  <cv>Se</cv> $3\leq5$ <cv>então</cv> $\dfrac{3}{5}\leq1$.

Poderíamos também ler a condicional $p\land q$ da seguinte forma: 
- $\dfrac{3}{5}\leq1$ é condição suficiente para $3\leq5$
- $3\leq5$ é condição necessária para $\dfrac{3}{5}\leq1$

Como exercício, podemos checar se a nossa $p\land q$ é <cg>$\,verdadeira\,$</cg>, para isso vamos tentar provar que: $(q\;$<cg>$verdadeira$</cg>$)$ é <cv>condição suficiente</cv> para $(p$\;ser\;$<cg>$verdadeira$</cg>$)$. Para isso partiremos de $p$ e utilizamndo apenas operações matemáticas permitidas tentaremos chegar em $q$.

$3\leq5$

Dividindo os dois lados por $5$, temos:

$\dfrac{3}{5}\leq\dfrac{5}{5}$

Como $\frac{5}{5}=1$, temos:

$\dfrac{3}{5}\leq1$

Na minha experiência, a melhor forma de não se perder nas condicionais é pensar em como você provaria que a condicional $r\land s$ é <cr>$falsa$</cr>. Se for impossível provar sua $\textcolor{red}{falsidade}$, ela só pode ser <cg>$\,verdadeira\,$</cg>. 

- $\quad\;\; r$ : É carnaval.
- $\quad\;\; s$ : Estou no bloquinho "sai de baixo".
- $r\land s$ : **Se** é carnaval **então** estou no bloquinho "sai de baixo".

- Se for um dia de carnaval ($p\;\textcolor{green}{V}$) e você me encontrar no bloquinho ($q\;\textcolor{green}{V}$) a condicional é <cg>$\,verdadeira\,$</cg>. "Eu falei a verdade"
- Se for um dia de carnaval ($p\;\textcolor{green}{V}$) e você **não** me encontrar no bloquinho ($q\;\textcolor{red}{F}$) a condicional é necessariamente <cr>$falsa$</cr>. "Eu estava mentindo"
- Se **não** for um dia de carnaval ($p\;\textcolor{red}{F}$), é impossível provar que a condicional é <cr>$falsa$</cr>. Pense no "se...então", como eu provo que a frase é <cr>$falsa$</cr> se já partimos do pressuposto de que a condição $p$ é <cr>$falsa$</cr>, não tem como. No nosso exemplo a única maneira de provar a $\textcolor{red}{falsidade}$ $seria ir ao bloquinho **em um dia de carnaval** e não me encontrar lá, para isso, é necessário que seja um dia de carnaval ($p\;\textcolor{green}{V}$)   

Seguindo essa lógica, preencheremos a tabela com <cr>$falso$</cr> somente quando a condição for <cg>$\,verdadeira\,$</cg> mas a conclusão $\textcolor{red}{não}$, ou seja, só a linha $\textcolor{green}{V}\;\textcolor{red}{F}$ será <cr>$\,F\,$</cr>

#### Tabela verdade
| $p$ | $q$ | $p \rightarrow q$ |
|:---:|:---:|:-----------------:|
| <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg>
| <cg>$\,V\,$</cg> | <cr>$\,F\,$</cr> | <cr>$\,F\,$</cr> 
| <cr>$\,F\,$</cr> | <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg>
| <cr>$\,F\,$</cr> | <cr>$\,F\,$</cr> | <cg>$\,V\,$</cg>

*Como preencher a tabela:*
- $p \rightarrow q$ só é <cr>$\,F\,$</cr>$</cg>extcolor{red}{F}$. Ou seja, basta procurar todas as linhas $\textcolor{green}{V}\;\textcolor{red}{F}$. 
- Todas as outras linhas são <cg>$\,V\,$</cg>

### Bicondicional:   $p \leftrightarrow q$

Podemos definir a bicondicional como sendo as duas condicionais ao mesmo tempo $(p \rightarrow q)$ e $(p \leftarrow q)$

A ideia dessa condicional é ser uma setinha de mão dupla, ela vai $(p \rightarrow q)$ e volta $(p \leftarrow q)$. $se p então q" e "se q então p" ao mesmo tempo.

> *lê-se: se p então q e se q então p" ou "se p então q e vice versa" ou "p se e somente se q"*

$p$ <cp>$\leftrightarrow$</cp> $q: (p $<cp>$\rightarrow$</cp> $q)$<cp>$\land$</cp>$(q $<cp>$\rightarrow$</cp> $p)$

$<cp>$se$</cp>$\; p\;$ <cp>$então$</cp>$\; q\;$<cp>$\;e\; vice\; versa$</cp>

$p\;$ <cp>se\;e\;somente\;se$</cp>\; $q$

Note que em todas as formas de leitura aparece um <cp>$e$</cp> em algum lugar, o que condiz muito com a existência do nosso conectivo <cp>$\land$</cp> na definição.

Dadas as frases anteriores da forma $p \rightarrow q$ com $p$ $q$ <cg>$\,verdadeiras\,$</cg> quais permanecem $p \leftrightarrow q$ <cg>$\,verdadeiras\,$</cg> quando transformadas em bicondicionais?

1. Ter mais de 18 anos é condição para comprar bebidas alcóolicas.
    - $\quad\quad\; p_{1}$ : Ter mais de 18 anos - assumimos ser <cg>$\,V\,$</cg>
    - $\quad\quad\; q_{1}$ : Poder comprar bebidas alcóolicas.
    - $p_{1} \rightarrow q_{1}$ : Se tenho mais de 18 anos então posso comprar bebidas alcóolicas.
        - Pela definição de condicional, ela só pode ser falsa quando a condição $p$ for <cg>$\,verdadeira\,$</cg> mas a conclusão for <cr>$falsa$</cr>. Como sabemos que todos que tem mais de 18 anos (condição $p\;$<cg>$V$</cg>) podem comprar bebidas alcóolicas (conclusão $q\;$<cg>$V$</cg>), a condicional é necessáriamente <cg>$\,verdadeira\,$</cg>.
    - $q_{1} \rightarrow p_{1}$ : Se posso comprar bebidas alcóolicas então tenho mais de 18 anos - <cg>$\,V\,$</cg>
        - O mesmo raciocínio pode ser aplicado aqui. Se uma certa pessoa pode comprar bebidas alcóolicas então ela tem mais de 18 anos, visto que ninguém venderia bebidas para um menor de idade. 
    - Como ambas ida e volta são <cg>$\,verdadeiras\,$</cg>, $(p_{1} \leftrightarrow q_{1})$ é <cg>$\,verdadeira\,$</cg>

Repita o raciocínio acima e avalie se as bicondicionais criadas a partir das condicionais a baixo são <cg>$\,verdadeiras\,$</cg>:

1. Estar grávida é condição para usar o assento preferencial em transportes públicos.
1. Estar matriculado no CPE é condição suficiente para assistir as aulas do cursinho.
1. Ter asas é condição para poder voar.

|:---:|:---:|:---------------------:|
| <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> |
| <cg>$\,V\,$</cg> | <cr>$\,F\,$</cr> | <cr>$\,F\,$</cr> |
| <cr>$\,F\,$</cr> | <cg>$\,V\,$</cg> | <cr>$\,F\,$</cr> |
| <cr>$\,F\,$</cr> | <cr>$\,F\,$</cr> | <cg>$\,V\,$</cg> |

*Como preencher a tabela:*
- $p \leftrightarrow q$ só é <cg>$\,verdadeiro\,$</cg> quando ambos $p$ e $q$ são <cg>$\,verdadeiros\,$</cg> ou ambos são <cr>$falsos$</cr>. Ou seja, basta procurar todas as linhas <cg>$\,V\,<cg>$\,V\,$</cg>\,$</cg> ou <cr>$\,F\,F\,$</cr>. 
- Todas as outras linhas são <cr>$falsas$</cr>

### Proposições sempre verdadeiras e sempre falsas

Algumas composições de proposições serão sempre <cg>$\,verdadeiras\,$</cg> (ou sempre <cr>$falsas$</cr>), independentemente do conteúdo das mesmas. Isso acontece não porque escolho acreditar que sempre são <cg>$\,verdadeiras\,$</cg> (ou sempre <cr>$falsas$</cr>), mas sim por conta das regras regras de composição que vimos até agora. 

#### Exemplos:

##### $(\sim p) \land p$

| $p$ | $\sim p$ | $(\sim p) \land p$ |
|:---:|:--------:|:---------------:|
| <cg>$\,V\,$</cg> | <cr>$\,F\,$</cr> | <cr>$\,F\,$</cr> |
| <cr>$\,F\,$</cr> | <cg>$\,V\,$</cg> | <cr>$\,F\,$</cr> |

Como a nossa tabela representa todas as combinações possíveis de $p$ e $\sim p$, temos que $\sim p \land p$ sempre é <cr>$falsa$</cr>, independente de quem $p$ seja.
- $\;\;\quad\quad p$ : Ventiladores ventilam
- $\;\;\quad\sim p$ : Ventiladores não ventilam
- $\sim p \land p$ : Ventiladores ventilam e não ventilam.

##### $(\sim p) \lor p$

| $p$ | $\sim p$ | $(\sim p) \lor p$ |
|:---:|:--------:|:--------------:|
| <cg>$\,V\,$</cg> | <cr>$\,F\,$</cr> | <cg>$\,V\,$</cg> |
| <cr>$\,F\,$</cr> | <cg>$\,V\,$</cg> | <cg>$\,V\,$</cg> |

Como a nossa tabela representa todas as combinações possíveis de $p$ e $\sim p$, temos que $\sim p \land p$ sempre é <cg>$\,verdadeira\,$</cg>, independente de quem $p$ seja.
- $\;\;\quad\quad p$ : Ventiladores ventilam
- $\;\;\quad\sim p$ : Ventiladores não ventilam
- $\sim p \lor p$ : Ventiladores ventilam ou não ventilam.


Proposições sempre verdadeiras, também são conhecidas como ***tautologias*** e serão muito úteis para definir equivalências.
