## Lógica

### Conjunção:$\quad p \land q$


> O nome conjunção vem do fato de que para $p \land q$ ser $\textcolor{green}{verdadeira}$ preciso que $p$ e $q$ sejam em conjunto, ao mesmo tempo, $\textcolor{green}{verdadeiras}$. Ideia de união, preciso das duas.

Elas juntas definem a conjunção. Garantir que só uma das duas proposições é $\textcolor{green}{verdadeira}$, não é suficiente para concluir que $p \land q$ também será. Mas se ambas forem $\textcolor{green}{verdadeiras}$, em conjunto, é possível garantir que a conjunção $p \land q$ também será $\textcolor{green}{verdadeira}$.

- *lê-se: "p e q"*
- $p \textcolor{pink}{\land} q\; :\; p\; \textcolor{pink}{e}\; q$

#### Exemplos:
- $\quad\;\; p$ : Meu é lanche gostoso.
- $\quad\;\; q$ : Meu é lanche bonito.
- $p\land q$ : Meu lanche é bonito **e** meu lanche é gostoso.

Aqui poderíamos reduzir a expressão ainda mais e dizer apenas "Meu lanche é bonito **e** gostoso". Nem sempre essa redução é simples

O lanche que comprei só pode ser bonito e gostoso ($p\land q$ $\textcolor{green}{verdadeira}$) se ao mesmo tempo $-$ conjuntamente $-$ ele for gostoso ($p$ $\textcolor{green}{verdadeira}$) e bonito ($q$ $\textcolor{green}{verdadeira}$).

- $\quad\;\; r$ : Quando está frio eu tomo banho quente.
- $\quad\;\; s$ : Quando tomo banho quente eu canto no chuveiro.
- $r\land s$ : Quando está frio eu tomo banho quente **e** quando tomo banho quente canto no chuveiro.

Aqui poderíamos reduzir para "Quando está frio eu tomo banho quente **e** canto no chuveiro"

Note aqui que se $r$ for $\textcolor{red}{falsa}$ a proposição $r\land s$ também será $\textcolor{red}{falsa}$ independente do valor de $s$. O mesmo raciocínio vale para $s$.
- Se $r$ é $\textcolor{red}{falsa}$, significa que em algum dia frio eu tomei um banho não quente. Note que nem preciso olhar para o restante da frase para conluir que a frase "Quando está frio eu tomo banho bem quente **e** canto no chuveiro", $r \land s$, é $\textcolor{red}{falsa}$.

#### Tabela verdade
| $p$ | $q$ | $p \land q$ |
|:---:|:---:|:-----------:|
| $\textcolor{green}{V}$ | $\textcolor{green}{V}$ | $\textcolor{green}{V}$ |
| $\textcolor{green}{V}$ | $\textcolor{red}{F}$ | $\textcolor{red}{F}$ |
| $\textcolor{red}{F}$ | $\textcolor{green}{V}$ | $\textcolor{red}{F}$ |
| $\textcolor{red}{F}$ | $\textcolor{red}{F}$ | $\textcolor{red}{F}$ |

*Como preencher a tabela:*
1. Escrevo todas as possíveis combinações de $p$ e $q$ : $\textcolor{green}{V}\textcolor{green}{V}$, $\textcolor{green}{V}\textcolor{red}{F}$, $\textcolor{red}{F}\textcolor{green}{V}$ e $\textcolor{red}{F}\textcolor{red}{F}$.
1. $p \textcolor{pink}{\land} q$ só é $\textcolor{green}{Verdadeiro}$ quando ambos $p$ $\textcolor{pink}{e}$ $q$ forem $\textcolor{green}{Verdadeiros}$. Ou seja, só preencho $\textcolor{green}{V}$ para a coluna $p \textcolor{pink}{\land} q$  na linha $\textcolor{green}{V V}$.
1. Todas as outras linhas são $\textcolor{red}{Falsas}$

### Disjunção:   $p \lor q$


> O nome disjunção vem do fato de que para $p \lor q$ ser $\textcolor{green}{verdadeira}$ basta que $p$ e $q$ sejam, separadamente, $\textcolor{green}{verdadeiras}$. Ideia de desunião, basta uma.

Enquanto para $``p\; e\; q"$ ser verdadeira, $p$ e $q$ precisavam ser, *juntamente*, $\textcolor{green}{verdadeiras}$; para $``p\; ou\; q"$ ser $\textcolor{green}{verdadeira}$, basta uma das duas, "separadamente"  ser $\textcolor{green}{verdadeira}$.

*lê-se: "p ou q"*

$p \textcolor{pink}{\lor} q:\; p\; \textcolor{pink}{ou}\; q$

#### Exemplos
- $\quad\;\; p$ : Vou ao cinema de metrô.
- $\quad\;\; q$ : Vou ao cinema de ônibus.
- $p\land q$ : Vou ao cinema de metrô ou vou ao cinema de ônibus.

Aqui poderíamos reduzir a disjunção para "Vou ao cinema de metrô ou ônibus". Note que essa frase continua $\textcolor{green}{V}$ mesmo quando eu vou só de metrô ($p$ $\textcolor{green}{V}$ e $q$ $\textcolor{red}{F}$) ou só de ônibus ($p$ $\textcolor{red}{F}$ e $q$ $\textcolor{green}{V}$). Basta que uma delas seja $\textcolor{green}{verdadeira}$ para a disjunção $p\lor q$ ser $\textcolor{green}{verdadeira}$.

#### Tabela verdade
| $p$ | $q$ | $p \lor q$ |
|:---:|:---:|:----------:|
| $\textcolor{green}{V}$ | $\textcolor{green}{V}$ | $\textcolor{green}{V}$ |
| $\textcolor{green}{V}$ | $\textcolor{red}{F}$ | $\textcolor{green}{V}$ |
| $\textcolor{red}{F}$ | $\textcolor{green}{V}$ | $\textcolor{green}{V}$ |
| $\textcolor{red}{F}$ | $\textcolor{red}{F}$ | $\textcolor{red}{F}$ |

*Como preencher a tabela:*
- $p \lor q$ só é $\textcolor{red}{F}$ quando ambos $p$ $\lor$ $q$ forem $\textcolor{red}{F}$. Ou seja, basta procurar todas as linhas $\textcolor{red}{F\;F}$. 
- Todas as outras linhas são $\textcolor{green}{V}$

Pela tabela verdade podemos observar que $p\lor q$ só $\textcolor{red}{F}$ quando ambas $p$ e $q$ forem simulataneamente $\textcolor{red}{F}$. Essa será a nossa leitura padrão.
- $p\land q$ só é $V$ na linha $\textcolor{green}{V\;V}$
- $p\lor q$ só é $\textcolor{red}{F}$ na linha $\textcolor{red}{F\;F}$

### Condicional:   $p \rightarrow q$

> O nome condicional já deixa claro que estamos falando de uma relação em que uma coisa é condição para outra. Se eu garantir uma certa coisa (condição) posso garantir uma outra coisa (conclusão).

- Ter mais de 18 anos é condição suficiente para poder comprar bebidas alcóolicas.
    - Se tenho sou maior de idade então posso comprar bebidas. 
- Estar grávida é condição suficiente para poder usar o assento preferencial em transportes públicos.
    - Se estou grávida então posso usar o assento preferencial.
- Estar matriculado no CPE é condição necessária para assistir as aulas do cursinho.
    - Se assisto aulas do cursinho então estou matriculado no CPE.
- Ter asas é condição necessária para poder voar.
    - Se posso voar então tenho asas.

Avaliando as frases acima, teríamos $\textcolor{green}{V\;V\;V\;}\textcolor{red}{F}$

*lê-se: "se p então q"*

$p \textcolor{pink}{\rightarrow} q:\; \textcolor{pink}{se}\; p\; \textcolor{pink}{então}\; q$

Existem também as seguintes leituras para $p \rightarrow q$ como explicitado nas frases acima.
- "p é condição suficiente para q"
- "q é condição necessária para p"

#### Exemplos

- $\quad\;\; p$ :  $3\leq5$
- $\quad\;\; q$ :  $\dfrac{3}{5}\leq1$
- $p\land q$ :  **Se** $3\leq5$ **então** $\dfrac{3}{5}\leq1$.

Poderíamos também ler a condicional $p\land q$ da seguinte forma: 
- $\dfrac{3}{5}\leq1$ é condição suficiente para $3\leq5$
- $3\leq5$ é condição necessária para $\dfrac{3}{5}\leq1$

Como exercício, podemos checar se a nossa $p\land q$ é $\textcolor{green}{verdadeira}$, para isso vamos tentar provar que: $(q\;\textcolor{green}{verdadeira})$ **é condição suficiente** para $(p$\;ser\;$\textcolor{green}{verdadeira})$. Para isso partiremos de $p$ e utilizamndo apenas operações matemáticas permitidas tentaremos chegar em $q$.

$3\leq5$

Dividindo os dois lados por $5$, temos:

$\dfrac{3}{5}\leq\dfrac{5}{5}$

Como $\frac{5}{5}=1$, temos:

$\dfrac{3}{5}\leq1$

Na minha experiência, a melhor forma de não se perder nas condicionais é pensar em como você provaria que a condicional $r\land s$ é $\textcolor{red}{falsa}$. Se for impossível provar sua $\textcolor{red}{falsidade}$, ela só pode ser $\textcolor{green}{verdadeira}$. 

- $\quad\;\; r$ : É carnaval.
- $\quad\;\; s$ : Estou no bloquinho "sai de baixo".
- $r\land s$ : **Se** é carnaval **então** estou no bloquinho "sai de baixo".

- Se for um dia de carnaval ($p\;\textcolor{green}{V}$) e você me encontrar no bloquinho ($q\;\textcolor{green}{V}$) a condicional é $\textcolor{green}{verdadeira}$. "Eu falei a verdade"
- Se for um dia de carnaval ($p\;\textcolor{green}{V}$) e você **não** me encontrar no bloquinho ($q\;\textcolor{red}{F}$) a condicional é necessariamente $\textcolor{red}{falsa}$. "Eu estava mentindo"
- Se **não** for um dia de carnaval ($p\;\textcolor{red}{F}$), é impossível provar que a condicional é $\textcolor{red}{falsa}$. Pense no "se...então", como eu provo que a frase é $\textcolor{red}{falsa}$ se já partimos do pressuposto de que a condição $p$ é $\textcolor{red}{falsa}$, não tem como. No nosso exemplo a única maneira de provar a $\textcolor{red}{falsidade}$ $seria ir ao bloquinho **em um dia de carnaval** e não me encontrar lá, para isso, é necessário que seja um dia de carnaval ($p\;\textcolor{green}{V}$)   

Seguindo essa lógica, preencheremos a tabela com $\textcolor{red}{falso}$ somente quando a condição for $\textcolor{green}{verdadeira}$ mas a conclusão $\textcolor{red}{não}$, ou seja, só a linha $\textcolor{green}{V}\;\textcolor{red}{F}$ será $\textcolor{red}{F}$ 

#### Tabela verdade
| $p$ | $q$ | $p \rightarrow q$ |
|:---:|:---:|:-----------------:|
| $\textcolor{green}{V}$ | $\textcolor{green}{V}$ | $\textcolor{green}{V}$ |
| $\textcolor{green}{V}$ | $\textcolor{red}{F}$ | $\textcolor{red}{F}$ |
| $\textcolor{red}{F}$ | $\textcolor{green}{V}$ | $\textcolor{green}{V}$ |
| $\textcolor{red}{F}$ | $\textcolor{red}{F}$ | $\textcolor{green}{V}$ |

*Como preencher a tabela:*
- $p \rightarrow q$ só é $\textcolor{red}{F}$ quando $p$ é $\textcolor{green}{V}$ e $q$ é $\textcolor{red}{F}$. Ou seja, basta procurar todas as linhas $\textcolor{green}{V}\;\textcolor{red}{F}$. 
- Todas as outras linhas são $\textcolor{green}{V}$

### Bicondicional:   $p \leftrightarrow q$

Podemos definir a bicondicional como sendo as duas condicionais ao mesmo tempo $(p \rightarrow q)$ e $(p \leftarrow q)$

A ideia dessa condicional é ser uma setinha de mão dupla, ela vai $(p \rightarrow q)$ e volta $(p \leftarrow q)$. $se p então q" e "se q então p" ao mesmo tempo.

> *lê-se: se p então q e se q então p" ou "se p então q e vice versa" ou "p se e somente se q"*

$p \textcolor{pink}{\leftrightarrow} q: (p \textcolor{pink}{\rightarrow} q)\textcolor{pink}{\land}(q \textcolor{pink}{\rightarrow} p)$

$\textcolor{pink}{se}\; p\; \textcolor{pink}{então}\; q\;\textcolor{pink}{\;e\; vice\; versa}$

$p\; \textcolor{pink}{se\;e\;somente\;se}\; q$

Note que em todas as formas de leitura aparece um $\textcolor{pink}{e}$ em algum lugar, o que condiz muito com a existência do nosso conectivo $\textcolor{pink}{\land}$ na definição.

Dadas as frases anteriores da forma $p \rightarrow q$ com $p$ $q$ $\textcolor{green}{verdadeiras}$ quais permanecem $p \leftrightarrow q$ $\textcolor{green}{verdadeiras}$ quando transformadas em bicondicionais?

1. Ter mais de 18 anos é condição para comprar bebidas alcóolicas.
    - $\quad\quad\; p_{1}$ : Ter mais de 18 anos - assumimos ser $\textcolor{green}{V}$
    - $\quad\quad\; q_{1}$ : Poder comprar bebidas alcóolicas.
    - $p_{1} \rightarrow q_{1}$ : Se tenho mais de 18 anos então posso comprar bebidas alcóolicas.
        - Pela definição de condicional, ela só pode ser falsa quando a condição $p$ for $\textcolor{green}{verdadeira}$ mas a $\textcolor{red}{conclusão}$ for falsa. Como sabemos que todos que tem mais de 18 anos ($p\;\textcolor{green}{V}$) podem comprar bebidas alcóolicas $(q\;\textcolor{green}{V}$), a condicional é necessáriamente $\textcolor{green}{verdadeira}$.
    - $q_{1} \rightarrow p_{1}$ : Se posso comprar bebidas alcóolicas então tenho mais de 18 anos - $\textcolor{green}{V}$
        - O mesmo raciocínio pode ser aplicado aqui. Se uma certa pessoa pode comprar bebidas alcóolicas então ela tem mais de 18 anos, visto que ninguém venderia bebidas para um menor de idade. 
    - Como ambas ida e volta são $\textcolor{green}{verdadeiras}$, $(p_{1} \leftrightarrow q_{1})$ é $\textcolor{green}{verdadeira}$

Repita o raciocínio acima e avalie se as bicondicionais criadas a partir das condicionais a baixo são $\textcolor{green}{verdadeiras}$:

1. Estar grávida é condição para usar o assento preferencial em transportes públicos.
1. Estar matriculado no CPE é condição suficiente para assistir as aulas do cursinho.
1. Ter asas é condição para poder voar.


Tabela verdade
| $p$ | $q$ | $p \leftrightarrow q$ |
|:---:|:---:|:---------------------:|
| $\textcolor{green}{V}$ | $\textcolor{green}{V}$ | $\textcolor{green}{V}$ |
| $\textcolor{green}{V}$ | $\textcolor{red}{F}$ | $\textcolor{red}{F}$ |
| $\textcolor{red}{F}$ | $\textcolor{green}{V}$ | $\textcolor{red}{F}$ |
| $\textcolor{red}{F}$ | $\textcolor{red}{F}$ | $\textcolor{green}{V}$ |

*Como preencher a tabela:*
- $p \leftrightarrow q$ só é $\textcolor{green}{verdadeiro}$ quando ambos $p$ e $q$ são $\textcolor{green}{verdadeiros}$ ou ambos são $\textcolor{red}{falsos}$. Ou seja, basta procurar todas as linhas $\textcolor{green}{V V}$ ou $\textcolor{red}{F F}$. 
- Todas as outras linhas são $\textcolor{red}{falsas}$

### Proposições sempre verdadeiras e sempre falsas

Algumas composições de proposições serão sempre $\textcolor{green}{verdadeiras}$ (ou sempre $\textcolor{red}{falsas}$), independentemente do conteúdo das mesmas. Isso acontece não porque escolho acreditar que sempre são $\textcolor{green}{verdadeiras}$ (ou sempre $\textcolor{red}{falsas}$), mas sim por conta das regras regras de composição que vimos até agora. 

#### Exemplos:

##### $\sim p \land p$

| $p$ | $\sim p$ | $\sim p \land p$ |
|:---:|:--------:|:---------------:|
| $\textcolor{green}{V}$ | $\textcolor{red}{F}$ | $\textcolor{red}{F}$ |
| $\textcolor{red}{F}$ | $\textcolor{green}{V}$ | $\textcolor{red}{F}$ |

Como a nossa tabela representa todas as combinações possíveis de $p$ e $\sim p$, temos que $\sim p \land p$ sempre é $\textcolor{red}{falsa}$, independente de quem $p$ seja.
- $\;\;\quad\quad p$ : Ventiladores ventilam
- $\;\;\quad\sim p$ : Ventiladores não ventilam
- $\sim p \land p$ : Ventiladores ventilam e não ventilam.

##### $\sim p \lor p$

| $p$ | $\sim p$ | $\sim p \lor p$ |
|:---:|:--------:|:--------------:|
| $\textcolor{green}{V}$ | $\textcolor{red}{F}$ | $\textcolor{green}{V}$ |
| $\textcolor{red}{F}$ | $\textcolor{green}{V}$ | $\textcolor{green}{V}$ |

Como a nossa tabela representa todas as combinações possíveis de $p$ e $\sim p$, temos que $\sim p \land p$ sempre é $\textcolor{green}{verdadeira}$, independente de quem $p$ seja.
- $\;\;\quad\quad p$ : Ventiladores ventilam
- $\;\;\quad\sim p$ : Ventiladores não ventilam
- $\sim p \lor p$ : Ventiladores ventilam ou não ventilam.


Proposições sempre verdadeiras, também são conhecidas como ***tautologias*** e serão muito úteis para definir equivalências. 

> Duas proposições $p$ e $q$ serão ditas equivalentes, isto é, falam a mesma coisa com palavras diferentes, se e somente se $p \leftrightarrow q$ for uma tautologia (só tiver $\textcolor{green}{V}$ na tabela). Para representar equivalências usaremos a notação $p \Leftrightarrow q$.
